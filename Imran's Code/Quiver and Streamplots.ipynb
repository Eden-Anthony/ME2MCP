```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(6, 1)
fig.set_figwidth(10)
fig.set_figheight(30)

#a

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

x, y = np.meshgrid(x,y)

u = x
v = y

ax1.quiver(x,y, u, v);
ax2.streamplot(x,y,u,v);

#b
u = y
v = -x
ax3.quiver(x,y,u,v);
ax4.streamplot(x,y,u,v);

#c
u = y * ((x**2+y**2))
v = -x * ((x**2)+y**2)
ax5.quiver(x,y,u,v);
ax6.streamplot(x,y,u,v);
```
