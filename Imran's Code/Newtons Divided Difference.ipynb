```python
def NewtDivDiff(xn,yn):
    
    n = len(xn) - 1
    
    #if we have one node then f[x0] = f(x0)
    if n == 0:
        f =  yn[0]
        
    elif n == 1:
        
        f = (yn[1] - yn[0])/(xn[1]-xn[0])
    
    else:
        #can use recursive function here
        f = (NewtDivDiff(xn[1:], yn[1:]) - NewtDivDiff(xn[:-1], yn[:-1])) / (xn[-1] - xn[0])
        
    
    return f

def NewtonInterp(xn,yn,x):
    
    #degree of polynomial will be one less than the amount of nodes
    k = len(xn) - 1
    
    y = []
    
    for point in range(len(x)):
    
        p = yn[0]
        
        for i in range(1,k+1):
            m = 1
            for j in range(i):
                m *= x[point] - xn[j]

            p += m * NewtDivDiff(xn[0:i+1], yn[0:i+1])
            

        y += [p]
        
    return y            
```

```python
#test a

import numpy as np
import matplotlib.pyplot as plt

dx = 0.05
xa = np.arange(0,3+dx,dx)
ya = np.sin(xa)

#two known nodes
xn = np.linspace(1,2,2)
yn = np.sin(xn)
print(yn)

y = NewtonInterp(xn,yn,xa)
y = np.array(y)

plt.plot(xa,y)
plt.plot(xa,ya)
plt.scatter(xn,yn);
```
