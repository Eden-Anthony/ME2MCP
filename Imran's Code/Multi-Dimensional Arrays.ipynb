```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
from mpl_toolkits import mplot3d
```

## 2-D Multi-Dimensional Arrays

```python
dx = 0.1
dy = 0.1

#creating arrays for domains
x = np.arange(-2*np.pi, 2*np.pi+dx,dx)
y = np.arange(-np.pi, 2*np.pi+dx,dx)

#creating a meshgrid to create all possible coordinates
(Xg, Yg) = np.meshgrid(x, y)

#define the respective functions
f = np.sin(Xg) * np.cos(Yg)
g = np.cos(Xg) * np.sin(Yg)

#surface plot
#defining a subplot with 1 row and 2 columns
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 2, 1, projection='3d')

ax.plot_surface(Xg,Yg,f)
ax.plot_wireframe(Xg,Yg,f)

ax = fig.add_subplot(1, 2, 2, projection='3d')

ax.plot_surface(Xg,Yg,g)
ax.plot_wireframe(Xg,Yg,g);

#fig, (ax3, ax4) = plt.subplots(1, 2)


#ax3.contour(Xg,Yg,f)
#ax4.contour(Xg,Yg,g);
```

## 3-D Multi-Dimensional Arrays

```python
#define time domain
dt = 0.05
t = np.arange(0,10+dt,dt)


(Xg, Yg, t) = np.meshgrid(x, y, t)

r = (np.sin(Xg) * np.cos(Yg)) * np.exp(-0.5*t)

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(1, 2, 1, projection='3d')

#plot snapshot of function when t = 0
ax.plot_surface(Xg[:,:,0],Yg[:,:,0],r[:,:,0])

ax = fig.add_subplot(1, 2, 2, projection='3d')
#plot snapshot of function when t=10
ax.plot_surface(Xg[:,:,0],Yg[:,:,0],r[:,:,-1]);
```
